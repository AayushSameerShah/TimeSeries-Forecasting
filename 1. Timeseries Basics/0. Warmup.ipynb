## Exercise `4`
*Generate the normal distribution (as done in the first ex) but here it should be on **2D**. Meaning, multivariate normal.*

```python
X = np.random.multivariate_normal([0, 0], [[1, -0.5], [-0.5, 2]], 1000)
plt.scatter(X[:, 0], X[:, 1]);
```

###### 

## Exercise `5`
*Calculate the sample `mean` and `covariance` of the data by **manual** method. (Not using np.mean() etc.)*

```python
x = X[:, 0]
y = X[:, 1]
```

##### Means

```python
mean_x = x.sum() / len(x)
mean_y = y.sum() / len(y)

mean_x, mean_y
```

##### Covariance

$\text{cov} = \frac{\sum{(x - \bar x)(y - \bar y)}}{n}$

```python
cov = ((x - mean_x) * (y - mean_y)).sum() / len(x)
cov
```

```python
# variance
var_x = ((x - mean_x) ** 2).sum() / len(x)
var_y = ((y - mean_y) ** 2).sum() / len(x)
```

Final Matrix

```python
cov_matrix = [[var_x, cov], [cov, var_y]]
cov_matrix
```

```python
# This is what np.cov() returns
np.cov(X.T)
```
