## I = integrated
This is the last component to study in the full ARIMA model.
- This integrated part basically ***takes difference*** of the previous values in the time-series.
- To do what? - To make the time-series become ***stationary***.

## üçΩ So basically

We have the time series: <br>
$\text{Given} = y_1, y_2, y_3, \cdots, y_t$

From differencing: <br>
$\text{Differenced} = (y_2 - y_1), (y_3 - y_2), (y_4 - y_3), \cdots, (y_t - y_{t-1})$

Where each differencing can be called: <br>
$\Delta y_t = y_t - y_{t-1}$

## üß† Why difference?

See, in the upcoming notebook for the ARIMA model we will *also* see the **ARMA** model. Meaning, just applying both AR & MA models without using the differencing. Quoted here because I wanted to use the term "ARMA" here. Don't worry about that right now, just know that there is something called ARMA as well.

Now,
- To run ARMA model... we want the data ***to be stationary***. 
- **Stationary**: The distribution does not change over time. 
- **Stationarity** is a nice property because it suggests the: *the mean, variance, auto-correlation etc. will be constant over time*.

*(for more discussion on this **"why?"** topic, please refer the last portion of `5. Stationarity` book in this section.)*

## üí° New thingk
*There is no typo (gk).*

It is that, in machine learning the *data is assumed to come from the same distribution*. It is good, because then and only then we can **get the pattern** in the data and **learn something** about the data.

If the data that we ***consider*** coming from the same distribution ***but*** is in reality from different distributions, then we will not learn anything useful!
<br>‚Äî<br>
In time-series, each new data point is a new position in time might be affected by many other things and making the overall time-series non-stationary.

> We are **stationarizing** the time-series because ***to make it come from*** the same distribution!

## üë®‚Äçüíª Some Code? Yes please!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20, 10)
```

```python
df = pd.read_csv("../data/sp500_close.csv", index_col=0, parse_dates=True)
aapl = df["AAPL"].dropna()
aapl.head()
```

There, we simply have the APPL stock data. Now, let's see how does that look.

```python
aapl.plot()
plt.title("The simple AAPLE data without any differencing used");
```

```python
aapl.diff(1).plot()
plt.title("The $1^{st}$ order difference");
```

```python
aapl.diff(2).plot()
plt.title("The $2^{nd}$ order difference");
```

```python
aapl.diff(1).plot(label="$1^{st}$ order")
aapl.diff(2).plot(label="$2^{nd}$ order")
plt.title("The $1^{st}$ & $2^{nd}$ order difference")
plt.legend();
```

Not much difference, huh?

```python
print("ùúá:", aapl.diff(1).mean())
print("ùúé:", aapl.diff(1).std())
```

```python
print("ùúá:", aapl.diff(2).mean())
print("ùúé:", aapl.diff(2).std())
```

Clearly, the first order difference is better as it has mean closer to 0 respectively and the standard deviation is low as well.
